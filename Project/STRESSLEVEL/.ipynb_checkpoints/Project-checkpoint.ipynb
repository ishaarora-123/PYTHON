```python
import tkinter as tk
from tkinter import messagebox, ttk
import mysql.connector
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import seaborn as sns
# Predefined username and password for login
VALID_USERNAME = "root"
VALID_PASSWORD = "admin123"

def connect_to_db():
    return mysql.connector.connect(
        host="localhost",  
        user="root",      
        password="admin123",  
        database="hobbiesnew",  
        auth_plugin="mysql_native_password"
    )
def fetch_data_as_dataframe():
    connection = connect_to_db()
    if connection:
        try:
            cursor = connection.cursor()
            cursor.execute("SELECT * FROM HobbiesSurvey")  # Replace with your table name
            rows = cursor.fetchall()  # Fetch all records from the table
            columns = [desc[0] for desc in cursor.description]  # Get column names from cursor
            connection.close()

            # Convert fetched data into a Pandas DataFrame
            df = pd.DataFrame(rows, columns=columns)
            return df  # Return the DataFrame

        except mysql.connector.Error as err:
            messagebox.showerror("Database Error", f"Error: {err}")
            return None
    else:
        messagebox.showerror("Connection Error", "Failed to connect to the database.")
        return None

def open_dashboard(): 
    global admin_window 
    admin_window = tk.Toplevel(root)
    admin_window.title("Admin Dashboard")
    admin_window.geometry("600x400")
    tk.Label(admin_window, text="Admin Dashboard", font=("Arial", 16)).pack(pady=20)
    tk.Button(admin_window, text="Show Database", command=show_database).pack(pady=10)
    tk.Button(admin_window, text="Delete Record", command=delete_database).pack(pady=10)
    tk.Button(admin_window, text="Display Visualizaatiosn", command=display_visualization).pack(pady=10)
def show_database():
    connection = connect_to_db()
    if connection:
        try:
            cursor = connection.cursor()
            cursor.execute("SELECT * FROM HobbiesSurvey")  # Replace with your table name
            rows = cursor.fetchall()  # Fetch all records from the table
            columns = [desc[0] for desc in cursor.description]  # Get column names from cursor
            # Create a new window to show the data
            show_window = tk.Toplevel(root)
            show_window.title("Database Records")
            # Create a Treeview widget to display the records
            tree = ttk.Treeview(show_window, columns=columns, show="headings")
            # Add the column headers to the Treeview
            for col in columns:
                tree.heading(col, text=col)  # Set heading
                tree.column(col, anchor=tk.CENTER)  # Center-align columns
            # Insert the rows of data into the Treeview
            for row in rows:
                tree.insert("", "end", values=row)
            # Pack the Treeview widget into the window
            tree.pack(expand=True, fill=tk.BOTH, padx=20, pady=20)
        except mysql.connector.Error as err:
            messagebox.showerror("Database Error", f"Error: {err}")
        finally:
            connection.close()
def delete_database():
    delete_window = tk.Toplevel(root)
    delete_window.title("Delete Record")
    tk.Label(delete_window, text="Enter ID to delete:").pack(pady=10)
    id_entry = tk.Entry(delete_window)
    id_entry.pack(pady=5)
    def delete_record():
        record_id = id_entry.get()
        if not record_id:
            messagebox.showerror("Error", "Please enter a record ID.")
            return
        connection = connect_to_db()
        if connection:
            try:
                cursor = connection.cursor()
                cursor.execute("DELETE FROM HobbiesSurvey WHERE id = %s", (record_id,))
                connection.commit()
                messagebox.showinfo("Success", "Record deleted successfully.")
                delete_window.destroy()
            except mysql.connector.Error as err:
                messagebox.showerror("Database Error", f"Error: {err}")
            finally:
                connection.close()
    tk.Button(delete_window, text="Delete", command=delete_record).pack(pady=10)
def display_visualization():
    visualization_window = tk.Toplevel(admin_window) 
    visualization_window.title("Visualizations")
    visualization_window.geometry("400x300")
    tk.Label(visualization_window, text="Choose a Visualization", font=("Arial", 14)).pack(pady=20)
    tk.Button(visualization_window, text="Hobby Distribution", command = create_hobby_distribution).pack(pady=10)
    tk.Button(visualization_window, text="Stress Level and Mental rate distribution", command = create_stress_distribution).pack(pady=10)
    tk.Button(visualization_window, text="Stress level and Social COntext Distribution", command=create_context_distribution).pack(pady=10)
def create_hobby_distribution():
    df = fetch_data_as_dataframe()
    if df is not None:
        hobby_counts = df['Frequency'].value_counts()
        fig, ax = plt.subplots(figsize=(6, 4))
        plt.pie(hobby_counts,labels = hobby_counts.index, autopct='%1.1f%%',colors=sns.color_palette('Set2'))
        ax.set_title('Hobby Distribution')
        ax.set_xlabel('Hobbies')
        ax.set_ylabel('Count')
        hobby_window = tk.Toplevel()
        hobby_window.title("Hobby Distribution")
        hobby_window.geometry("600x500")  # Set size for the graph window
        canvas = FigureCanvasTkAgg(fig, master=hobby_window)
        canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)
        canvas.draw()
def create_stress_distribution():
    df = fetch_data_as_dataframe()
    if df is not None:
        stress_level_rating = df.groupby(['StressLevel' , 'MentalHealth']).size().unstack()
        list1 = df['StressLevel'].unique()
        stress_level_rating = stress_level_rating.reindex(index=list1, columns=np.arange(1, 11), fill_value=0)
        X_axis = np.arange(1,11)
        fig, ax = plt.subplots(figsize=(10, 6))
        ax.bar(X_axis-0.25 ,stress_level_rating.loc['Low'],0.25, label = 'Low Stress',color = 'pink')
        ax.bar(X_axis ,stress_level_rating.loc['Moderate'],0.25, label = 'Moderate Stress',color = 'lightblue')
        ax.bar(X_axis+0.25 ,stress_level_rating.loc['High'],0.25, label = 'High Stress',color = 'lightgreen')
        ax.set_xticks(X_axis)
        y_max = stress_level_rating.max().max()
        ax.set_yticks(np.arange(0, y_max + 5, step=1))
        ax.set_xticklabels(X_axis)
        ax.set_xlabel('Mental Health Rating (1 to 10)')
        ax.set_ylabel('Number of People')
        ax.set_title('Mental Health Ratings by Stress Level')
        ax.legend()
        stress_window = tk.Toplevel()
        stress_window.title("Hobby Distribution")
        stress_window.geometry("600x500")  # Set size for the graph window
        canvas = FigureCanvasTkAgg(fig, master=stress_window)
        canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)
        canvas.draw()
def create_context_distribution():
    df = fetch_data_as_dataframe()
    if df is not None:
        type_of_engage_in_hobbies = df.groupby(['SocialContext' , 'StressLevel']).size().unstack()
        type_of_engage_in_hobbies = type_of_engage_in_hobbies.reindex(index=['With friends or groups','Alone','Both equally'])
        stress_levels = ['Low' , 'Moderate' , 'High']
        X_axis = np.arange(len(stress_levels))
        fig, ax = plt.subplots(figsize=(10, 6))
        ax.bar(X_axis-0.25 ,type_of_engage_in_hobbies.loc['With friends or groups'],0.25, label = 'With friends or groups',color = 'pink')
        ax.bar(X_axis ,type_of_engage_in_hobbies.loc['Alone'],0.25, label = 'Alone',color = 'lightblue')
        ax.bar(X_axis+0.25 ,type_of_engage_in_hobbies.loc['Both equally'],0.25, label = 'Both equally',color = 'lightgreen')
        ax.set_xticks(X_axis)
        ax.set_yticks(np.arange(0,20))
        ax.set_xticklabels(stress_levels)
        ax.set_xlabel('Type of engagement of the person in the hobbies')
        ax.set_ylabel('Number of People')
        ax.set_title('Type of Engagement with Current Stress Level')
        ax.legend()
        context_window = tk.Toplevel()
        context_window.title("Hobby Distribution")
        context_window.geometry("600x500")  # Set size for the graph window
        canvas = FigureCanvasTkAgg(fig, master=context_window)
        canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)
        canvas.draw()

def login():
    username = username_entry.get()
    password = password_entry.get()

    if username == VALID_USERNAME and password == VALID_PASSWORD:
        open_dashboard()
    else:
        connection = connect_to_db()
        if connection:
            try:
                cursor = connection.cursor()
                cursor.execute("SELECT * FROM Users WHERE username = %s AND password = %s", (username, password))
                result = cursor.fetchone()
                if result:
                    messagebox.showinfo("Login Success", "Welcome!")  # Close the login window after successful login
                    open_survey_form()  
                else:
                    messagebox.showerror("Login Failed", "Invalid username or password.")
            except mysql.connector.Error as err:
                messagebox.showerror("Database Error", f"Error: {err}")
            finally:
                connection.close()
        
def open_register_page():
    global register_window, reg_username_entry, reg_password_entry, reg_confirm_password_entry
    register_window = tk.Toplevel(root)
    register_window.title("Register")
    register_window.geometry("300x300")

    # Registration form fields
    tk.Label(register_window, text="Username").pack(pady=5)
    reg_username_entry = tk.Entry(register_window)
    reg_username_entry.pack(pady=5)

    tk.Label(register_window, text="Password").pack(pady=5)
    reg_password_entry = tk.Entry(register_window, show="*")
    reg_password_entry.pack(pady=5)

    tk.Label(register_window, text="Confirm Password").pack(pady=5)
    reg_confirm_password_entry = tk.Entry(register_window, show="*")
    reg_confirm_password_entry.pack(pady=5)

    tk.Button(register_window, text="Register", command=register_user).pack(pady=10)

def register_user():
    global reg_username_entry, reg_password_entry, reg_confirm_password_entry
    username = reg_username_entry.get()
    password = reg_password_entry.get()
    confirm_password = reg_confirm_password_entry.get()

    if not username or not password:
        messagebox.showerror("Error", "All fields are required!")
        return
    if password != confirm_password:
        messagebox.showerror("Error", "Passwords do not match!")
        return

    connection = connect_to_db()
    if connection:
        try:
            cursor = connection.cursor()
            cursor.execute("SELECT * FROM Users WHERE username = %s", (username,))
            if cursor.fetchone():
                messagebox.showerror("Error", "Username already exists!")
                return

            cursor.execute("INSERT INTO Users (username, password) VALUES (%s, %s)", (username, password))
            connection.commit()
            messagebox.showinfo("Success", "Registration successful!")
            register_window.destroy()
        except mysql.connector.Error as err:
            messagebox.showerror("Database Error", f"Error: {err}")
        finally:
            connection.close()

def submit_data():
    name = name_entry.get()
    
    age_group = age_var.get()
    
    gender = gender_var.get()
    
    hobbies = []
    if physical_var.get(): hobbies.append("Physical")
    if social_var.get(): hobbies.append("Social")
    if intellectual_var.get(): hobbies.append("Intellectual")
    if creative_var.get(): hobbies.append("Creative")
    if others_var.get(): hobbies.append("Others")
    hobbies_selected = ', '.join(hobbies)
    
    frequency = frequency_var.get()
    
    hours_per_week = hours_var.get()
    
    mental_health = mental_health_var.get()
    
    stress_level = stress_var.get()
    
    mood_effect = mood_var.get()
    
    benefits = []
    if relaxation_var.get(): benefits.append("Relaxation") 
    if reduced_stress_var.get(): benefits.append("Reduced Stress")
    if focus_var.get(): benefits.append("Improved Focus")
    if increased_var.get(): benefits.append("Increased Happiness")
    if no_var.get(): benefits.append("No Benefit")
    benefits_selected = ', '.join(benefits)
    
    social_context = context_var.get()
    
    try:
        connection = mysql.connector.connect(
            host='localhost',
            user='root',
            password='admin123',
            database='hobbiesnew',
            auth_plugin='mysql_native_password'
        )
        cursor = connection.cursor()
        cursor.execute(
            """
            INSERT INTO HobbiesSurvey 
            (Name, AgeGroup, Gender, Hobbies, Frequency, HoursPerWeek, MentalHealth, StressLevel, MoodEffect, Benefits, SocialContext)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
            """,
            (name, age_group, gender, hobbies_selected, frequency, hours_per_week, mental_health, stress_level, mood_effect, benefits_selected, social_context)
        )
        connection.commit()
        messagebox.showinfo("Submission Successful", "Thank you for your submission!")
        survey_window.destroy()
    except mysql.connector.Error as err:
        messagebox.showerror("Database Error", f"Error: {err}")
    finally:
        connection.close()

# Function to open the survey form
def open_survey_form():
    global survey_window
    survey_window = tk.Tk()
    survey_window.title("Hobbies Survey")
    survey_window.geometry("1100x600")

    # Add a canvas for scrolling
    canvas = tk.Canvas(survey_window)
    scrollbar = ttk.Scrollbar(survey_window, orient="vertical", command=canvas.yview)
    scrollable_frame = ttk.Frame(canvas)

    # Configure scroll behavior
    scrollable_frame.bind(
        "<Configure>",
        lambda e: canvas.configure(
            scrollregion=canvas.bbox("all")
        )
    )

    canvas.create_window((0, 0), window=scrollable_frame, anchor="nw")
    canvas.configure(yscrollcommand=scrollbar.set)
    
    canvas.pack(side="left", fill="both", expand=True)
    scrollbar.pack(side="right", fill="y")
    
    global name_entry, age_var, gender_var
    global physical_var, social_var, intellectual_var, creative_var, others_var
    global frequency_var, hours_var, mental_health_var, stress_var
    global mood_var, relaxation_var, reduced_stress_var, focus_var, context_var, increased_var, no_var

    tk.Label(scrollable_frame, text="Name").grid(row=0, column=0, sticky="w", padx=10, pady=5)
    name_entry = tk.Entry(scrollable_frame)
    name_entry.grid(row=0, column=1, padx=10, pady=5, sticky="w")
    
    tk.Label(scrollable_frame, text="What is your age group?").grid(row=1, column=0, sticky="w", padx=10, pady=5)
    age_var = tk.StringVar()
    tk.Radiobutton(scrollable_frame, text="14-17", variable=age_var, value="14-17").grid(row=1, column=1, sticky="w")
    tk.Radiobutton(scrollable_frame, text="18-24", variable=age_var, value="18-24").grid(row=1, column=2, sticky="w")
    tk.Radiobutton(scrollable_frame, text="25-34", variable=age_var, value="25-34").grid(row=1, column=3, sticky="w")
    tk.Radiobutton(scrollable_frame, text="35-44", variable=age_var, value="35-44").grid(row=1, column=4, sticky="w")
    tk.Radiobutton(scrollable_frame, text="45-54", variable=age_var, value="45-54").grid(row=1, column=5, sticky="w")
    
    tk.Label(scrollable_frame, text="Gender").grid(row=2, column=0, sticky="w", padx=10, pady=5)
    gender_var = tk.StringVar()
    tk.Radiobutton(scrollable_frame, text="Male", variable=gender_var, value="Male").grid(row=2, column=1, sticky="w")
    tk.Radiobutton(scrollable_frame, text="Female", variable=gender_var, value="Female").grid(row=2, column=2, sticky="w")
    
    tk.Label(scrollable_frame, text="Type of Hobbies you engage in?(Select all that apply)").grid(row=3, column=0, sticky="w", padx=10, pady=5)
    physical_var = tk.BooleanVar()
    social_var = tk.BooleanVar()
    intellectual_var = tk.BooleanVar()
    creative_var = tk.BooleanVar()
    others_var = tk.BooleanVar()
    tk.Checkbutton(scrollable_frame, text="Physical", variable=physical_var).grid(row=3, column=1, sticky="w")
    tk.Checkbutton(scrollable_frame, text="Social", variable=social_var).grid(row=3, column=2, sticky="w")
    tk.Checkbutton(scrollable_frame, text="Intellectual", variable=intellectual_var).grid(row=3, column=3, sticky="w")
    tk.Checkbutton(scrollable_frame, text="Creative", variable=creative_var).grid(row=3, column=4, sticky="w")
    tk.Checkbutton(scrollable_frame, text="Others", variable=others_var).grid(row=3, column=5, sticky="w")
    
    tk.Label(scrollable_frame, text="How often do you engage in your hobbies?").grid(row=4, column=0, sticky="w", padx=10, pady=5)
    frequency_var = tk.StringVar()
    tk.Radiobutton(scrollable_frame, text="Daily", variable=frequency_var, value="Daily").grid(row=4, column=1, sticky="w")
    tk.Radiobutton(scrollable_frame, text="Several times a week", variable=frequency_var, value="Several times a week").grid(row=4, column=2, sticky="w")
    tk.Radiobutton(scrollable_frame, text="Weekly", variable=frequency_var, value="Weekly").grid(row=4, column=3, sticky="w")
    tk.Radiobutton(scrollable_frame, text="Monthly", variable=frequency_var, value="Monthly").grid(row=4, column=4, sticky="w")
    tk.Radiobutton(scrollable_frame, text="Less frequently", variable=frequency_var, value="Less frequently").grid(row=4, column=5, sticky="w")
    
    tk.Label(scrollable_frame, text="On average, how many hours per week do you spend on your hobbies?").grid(row=5, column=0, sticky="w", padx=10, pady=5)
    hours_var = tk.StringVar()
    tk.Radiobutton(scrollable_frame, text="Less than 1 hour", variable=hours_var, value="<1").grid(row=5, column=1, sticky="w")
    tk.Radiobutton(scrollable_frame, text="1-3 hours", variable=hours_var, value="1-3").grid(row=5, column=2, sticky="w")
    tk.Radiobutton(scrollable_frame, text="4-7 hours", variable=hours_var, value="4-7").grid(row=5, column=3, sticky="w")
    tk.Radiobutton(scrollable_frame, text="8+ hours", variable=hours_var, value="8+").grid(row=5, column=4, sticky="w")

    tk.Label(scrollable_frame, text="On a scale from 1 to 10, how would you rate your overall mental health?").grid(row=6, column=0, sticky="w", padx=10, pady=5)
    mental_health_var = tk.IntVar()
    tk.Scale(scrollable_frame, from_=1, to=10, orient=tk.HORIZONTAL, variable=mental_health_var).grid(row=6, column=1, columnspan=3, sticky="w")

    tk.Label(scrollable_frame, text="What is your current stress level?").grid(row=7, column=0, sticky="w", padx=10, pady=5)
    stress_var = tk.StringVar()
    tk.Radiobutton(scrollable_frame, text="Low", variable=stress_var, value="Low").grid(row=7, column=1, sticky="w")
    tk.Radiobutton(scrollable_frame, text="Moderate", variable=stress_var, value="Moderate").grid(row=7, column=2, sticky="w")
    tk.Radiobutton(scrollable_frame, text="High", variable=stress_var, value="High").grid(row=7, column=3, sticky="w")
    
    tk.Label(scrollable_frame, text="How does engaging in your hobbies affect your mood?").grid(row=8, column=0, sticky="w", padx=10, pady=5)
    mood_var = tk.StringVar()
    tk.Radiobutton(scrollable_frame, text="Improves my mood", variable=mood_var, value="Improves my mood").grid(row=8, column=1, sticky="w")
    tk.Radiobutton(scrollable_frame, text="No noticeable change", variable=mood_var, value="No noticeable change").grid(row=8, column=2, sticky="w")
    tk.Radiobutton(scrollable_frame, text="Worsen my mood", variable=mood_var, value="Worsen my mood").grid(row=8, column=3, sticky="w")

    tk.Label(scrollable_frame, text="What benefits do you gain from engaging in your hobbies? (Select all that apply)").grid(row=9, column=0, sticky="w", padx=10, pady=5)
    relaxation_var = tk.BooleanVar()
    reduced_stress_var = tk.BooleanVar()
    focus_var = tk.BooleanVar()
    increased_var = tk.BooleanVar()
    no_var = tk.BooleanVar()
    tk.Checkbutton(scrollable_frame, text="Relaxation", variable=relaxation_var).grid(row=9, column=1, sticky="w")
    tk.Checkbutton(scrollable_frame, text="Reduced Stress", variable=reduced_stress_var).grid(row=9, column=2, sticky="w")
    tk.Checkbutton(scrollable_frame, text="Improved Focus", variable=focus_var).grid(row=9, column=3, sticky="w")
    tk.Checkbutton(scrollable_frame, text="Increased Happiness", variable=increased_var).grid(row=9, column=4, sticky="w")
    tk.Checkbutton(scrollable_frame, text="No Benefit", variable=no_var).grid(row=9, column=5, sticky="w")

    tk.Label(scrollable_frame, text="Do you mostly engage in your hobbies alone or with others?").grid(row=10, column=0, sticky="w", padx=10, pady=5)
    context_var = tk.StringVar()
    tk.Radiobutton(scrollable_frame, text="Alone", variable=context_var, value="Alone").grid(row=10, column=1, sticky="w")
    tk.Radiobutton(scrollable_frame, text="With friends", variable=context_var, value="With friends").grid(row=10, column=2, sticky="w")
    tk.Radiobutton(scrollable_frame, text="Both", variable=context_var, value="Both").grid(row=10, column=3, sticky="w")

    # Submit Button
    tk.Button(scrollable_frame, text="Submit", command=submit_data).grid(row=11, column=1, columnspan=3, pady=20)

    survey_window.mainloop()

root = tk.Tk()
root.title("Login")
root.geometry("400x300")

# Username label and entry
tk.Label(root, text="Username").pack(pady=10)
username_entry = tk.Entry(root)
username_entry.pack(pady=5)

# Password label and entry
tk.Label(root, text="Password").pack(pady=10)
password_entry = tk.Entry(root, show="*")
password_entry.pack(pady=5)
tk.Button(root, text="Login", command=login).pack(pady=10)
tk.Button(root, text="Register", command=open_register_page).pack(pady=10)

root.mainloop()
```

```python
import pandas as pd
import mysql.connector

# Step 1: Read the CSV file
csv_file_path = 'Responses.csv'
df = pd.read_csv(csv_file_path, encoding='ISO-8859-1')

# Step 3: Connect to MySQL
connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='admin123',
    database='hobbiesnew',
    auth_plugin='mysql_native_password'
)
cursor = connection.cursor()

# Step 4: Prepare SQL Query
table_name = 'HobbiesSurvey'
columns = ", ".join([f"`{col}`" for col in df.columns])  # Add backticks
placeholders = ", ".join(["%s"] * len(df.columns))       # Placeholders for values
insert_query = f"INSERT INTO {table_name} ({columns}) VALUES ({placeholders})"

# Step 5: Insert Rows into MySQL Table
for row in df.itertuples(index=False):
    cursor.execute(insert_query, row)

# Commit changes and close connection
connection.commit()
cursor.close()
connection.close()

print("Data inserted successfully!")
```
